```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

# Import our dataset
df = pd.read_csv('iris.csv', index_col=0)
df.head()
```

```python
# Replace all the values in the species colomns that are not Iris-setoa with 
# Not-Setosa
df.loc[df['Species'] != 'Iris-setosa', 'Species'] = 'Not-setosa'
df.tail()
```

```python
# Label Encode our Species data
Label_encoder = LabelEncoder()
df['Species'] = Label_encoder.fit_transform(df['Species'])
df.head()
```

```python
# Identify our independant variable X.
X = df.iloc[:,[0,1,2,3]].values
```

```python
# Encoding our y variable
y = df.iloc[:,4].values
```

```python
# Train our data 
X = X.reshape(-1, 4)
X = preprocessing.scale(X) # scale the data so that it is easier to fit

X_train, X_test, y_train, y_test = train_test_split(X, 
                y, test_size=0.25, random_state=0)
```

```python
# fit a model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# make predictions on test data
y_pred = log_reg.predict(X_test).reshape(-1,1)
```

```python
# Create a confusion matrix
from sklearn.metrics import confusion_matrix

classes = ['Species_Iris-setosa', 'Not_Setosa']
conf_mat = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(conf_mat, columns=classes, index=classes)
cm_df
```

```python
# From the results of our confusion matrix we can clearly see that
# our model has successfully predicted every single one of our 
# test data samples correctly which indicates a high level of accuracy.
# From this infomation we can assume that our data would have both 
# higher precision and recall. 
```
